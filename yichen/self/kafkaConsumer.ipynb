```python
from kafka import KafkaConsumer
consumer = KafkaConsumer(bootstrap_servers='localhost:9092',
                         auto_offset_reset='earliest',
                         group_id='test1',
                         consumer_timeout_ms = 1000
                        )
consumer.subscribe('tweets-kafka-test')
```

```python
for msg in consumer:
    print(msg)
```
